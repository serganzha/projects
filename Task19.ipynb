```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline

df = pd.read_csv("insurance.csv")

#defining values of x and y
x = df[["age"]].values
y = df["charges"].values
#plotting the scatter plot
plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

There is a definite positive correlation between age and insurance charges. However we need to consider the rest of the data provided, as they effect the charges parameter. More advanced modeling techniques that can handle multiple variables simultaneously, such as Multiple Linear Regression, would be better suited to demonstrate how factors like smoking or BMI for example, correlate with increased charges within the same age groups. 

```python
#fittign a linear regression model to the data, and a line of best fit
reg = LinearRegression()
reg.fit(x, y)
```

```python
#plotitng the scatter plot with the best-fit line
y_pred = reg.predict(x)

plt.scatter(x,y)
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel("Charges")
plt.title("Linear Regression")
plt.show()
```
